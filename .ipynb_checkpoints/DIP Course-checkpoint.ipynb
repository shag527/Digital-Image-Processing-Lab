```python
# Importing libraries

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def display(img,title="Original"):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img,cmap='gray'),plt.title(title)
    plt.show()
```

```python
im=cv2.imread('/home/shagun/Downloads/test.jpeg')
display(im)
```

```python
count=np.zeros((256,))
x,y,z=np.shape(im)
for k in range(z):
    for i in range(x):
        for j in range(y):
            count[im[i,j,k]]+=1

bins=[]
for i in range(0,255,5):
    bins.append(i)
            
fig, ax = plt.subplots(figsize =(10, 7)) 
ax.hist(count, bins = bins) 
```
