```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
im=cv2.imread('/home/shagun/Downloads/pic.jpeg')
```

```python
def display(img,title="Original"):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img,cmap='gray'),plt.title(title)
    plt.show()
```

```python
display(im)
```

# Mirroring

```python
# Flipping image vertically 
[x,y,z]=np.shape(im)

for i in range(z):
    for j in range(x):
        for k in range(int(y/2)):
            im[j,k,i],im[j,y-k-1,i]=im[j,y-k-1,i],im[j,k,i]

display(im,"Mirror Image")
```

# Rotation

```python
def moveBy90(width, i, j):
    iNew = -j + width - 1
    jNew = i
    return int(iNew), int(jNew)

image1 = cv2.imread('/home/shagun/Downloads/pic.jpeg')
image2 = np.zeros([image1.shape[1], image1.shape[0], image1.shape[2]], dtype=image1.dtype)
for k in range(z):
    for i in range(x):
        for j in range(y):
            newPoint = moveBy90(y, i, j)
            image2[newPoint[0], newPoint[1], k]=image1[i,j,k]
```

```python
display(image2,"Rotated Image")
```

# Enlargement

```python
originalImage = cv2.imread('/home/shagun/Downloads/pic.jpeg')
[x,y,z] = np.shape(originalImage);

newImage = np.zeros([int(x*2),int(y*2),z],dtype=originalImage.dtype);


for k in range(z):
    x_count=0
    for i in range(x):
        y_count=0
        for j in range(y):
            newImage[i+x_count,j+y_count,k]=originalImage[i,j,k]
            newImage[i+x_count,j+y_count+1,k]=originalImage[i,j,k]
            newImage[i+x_count+1,j+y_count,k]=originalImage[i,j,k]
            newImage[i+x_count+1,j+y_count+1,k]=originalImage[i,j,k]
            y_count+=1
        x_count+=1

display(newImage,"Enlarged Image")
```
