# Filtering Images

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
im=cv2.imread('/home/shagun/Downloads/hpf.jpeg')

def display(img,title="Original"):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img,cmap='gray'),plt.title(title)
    plt.show()
```

### Low Paas Filter

```python
def conv(img,filter):
    x,y,z=np.shape(img)
    fx,fy=np.shape(filter)
    low_img=np.zeros((x-fx+1,y-fy+1,z))
    
    for k in range(z):
        for i in range(x-fx+1):
            for j in range(y-fy+1):
                low_img[i,j,k]=np.sum(filter*img[i:i+fx,j:j+fy,k])
    
    return low_img.astype(np.uint8)
```

```python
(lkx,lky)=(3,3)
kernel=np.ones((lkx,lky))*1/(lkx*lky)

low_img=conv(im,kernel)
display(im)
display(low_img,"Low Pass Filter")
```

### High Pass Filter

```python
(hkx,hky)=(5,5)
kernel=-1/(hkx*hky)*np.ones((hkx,hky))
kernel[int(hkx/2),int(hky/2)]=-np.sum(kernel)-(1/(hkx*hky))

im=cv2.imread('/home/shagun/Downloads/dice.jpeg')
display(im)
high_img=conv(im,kernel)
display(high_img,"High Pass Filter")
```
