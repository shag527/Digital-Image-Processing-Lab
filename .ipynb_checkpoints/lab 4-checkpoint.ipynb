# Filtering 8 bit Images

```python
# Importing libraries

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def display(img,title="Original"):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img,cmap='gray'),plt.title(title)
    plt.show()
```

## Smoothing Color Image

```python
def conv(img,filter):
    x,y=np.shape(img)
    fx,fy=np.shape(filter)
    low_img=np.zeros((x-fx+1,y-fy+1))
    
    for i in range(x-fx+1):
        for j in range(y-fy+1):
            low_img[i,j]=np.sum(filter*img[i:i+fx,j:j+fy])
    
    return low_img.astype(np.uint8)
```

```python
im=cv2.imread('/home/shagun/Downloads/parrots.jpeg')
display(im)

(lkx,lky)=(5,5)
kernel=np.ones((lkx,lky))*1/(lkx*lky)

blue,green,red=cv2.split(im)

res_b=conv(blue,kernel)
res_g=conv(green,kernel)
res_r=conv(red,kernel)

res=cv2.merge((res_b,res_g,res_r))

display(res,"Smoothened Image")
```

## Sharpening Color Image

```python
im=cv2.imread('/home/shagun/Downloads/parrots.jpeg')
display(im)

(hkx,hky)=(5,5)
kernel=-1/(hkx*hky)*np.ones((hkx,hky))
kernel[int(hkx/2),int(hky/2)]=-np.sum(kernel)-(1/(hkx*hky))

blue,green,red=cv2.split(im)

res_b=conv(blue,kernel)
res_g=conv(green,kernel)
res_r=conv(red,kernel)

res=cv2.merge((res_b,res_g,res_r))

display(res,"Sharpened Image")
```
