## Image segmentation using OSTU and Thresholding Method

```python
# IMporting libraries

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Reading Image

im=cv2.imread('/home/shagun/Downloads/road.jpeg')

def display(img,title="Original"):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img,cmap='gray'),plt.title(title)
    plt.show()
    
display(im)
```

```python
# Converting image to grayscale

gray_im=cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
display(gray_im,"Gray Image")
```

### Using OTSU Algorithm

```python
plt.hist(gray_im.ravel(),256,[0,256])
plt.title("Image Histogram")
plt.show()
```

```python
# Implementing OTSU Algorithm

def otsu(image):
    total_pixels=image.shape[0]*image.shape[1]
    mean_weight=1.0/total_pixels
    hist,bins=np.histogram(image,np.array(range(0,256)))
    
    final_thresh,final_value=[-1,-1]
    for i in bins[1:-1]:
        w0=np.sum(hist[:i])*mean_weight
        w1=np.sum(hist[i:])*mean_weight
        
        mub=np.mean(hist[:i])
        muf=np.mean(hist[i:])
        
        value=w0*w1*(mub-muf)**2
        
        if value>final_value:
            final_thresh=i
            final_value=value
            
    res=image.copy()
    res[image>final_thresh]=255
    res[image<final_thresh]=0
    return res
```

```python
# Calling OTSU Function

res=otsu(gray_im)
display(res,"Segmentation using OTSU")
```

### Using Thresholding Method

```python
# Defining Threshold function

def threshold(image,thresh):
    res=image.copy()
    res[image>thresh]=255
    res[image<thresh]=0
    return res

res=threshold(gray_im,160)
display(res,"Segmentation using Thresholding")
```
