# Filtering Images

```python
# Importing libraries

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def display(img,title="Original"):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img,cmap='gray'),plt.title(title)
    plt.show()
```

## Arithmetic Mean Filter

```python
def arithmetic_filter(img,size):
    x,y,z=np.shape(img)
    res=np.zeros((x,y,z))
    idx=int(size/2);
    
    for k in range(z):
        for i in range(x-size):
            for j in range(y-size):
                res[i+idx,j+idx,k]=np.sum(img[i:i+size,j:j+size,k])/(size*size)
    
    return res.astype(np.uint8)
    
```

```python
im=cv2.imread('/home/shagun/Downloads/sap1.jpeg')

res_amf=arithmetic_filter(im,3)
display(im)
display(res_amf,"Arithmetic Mean Filter")
```

## Geometric Mean Filter

```python
def geometric_filter(img,size):
    x,y,z=np.shape(img)
    res=np.zeros((x,y,z))
    idx=int(size/2)
    
    for k in range(z):
        for i in range(x-size):
            for j in range(y-size):
                res[i+idx,j+idx,k]=(np.prod(img[i:i+size,j:j+size,k]))**(1/(size*size))
    
    return res.astype(np.uint8)
    
```

```python
im=cv2.imread('/home/shagun/Downloads/gauss.png')

res_gmf=geometric_filter(im,3)
display(im)
display(res_gmf,"Geometric Mean Filter")
```
