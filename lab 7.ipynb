## Line and Edge Detection

```python
# importing libraries

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
im=cv2.imread('/home/shagun/Downloads/image.png')

def display(img,title="Original Image"):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img,cmap='gray'),plt.title(title)
    plt.show()
    
display(im)
```

```python
# Converting image to grayscale

gray_im=cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
display(gray_im,"Gray Image")
```

```python
# Defining convolution function

def conv(img,sv,sh):
    x,y=np.shape(img)
    fx,fy=np.shape(sh)
    res=np.zeros((x-fx+1,y-fy+1))
    
    for i in range(x-fx+1):
        for j in range(y-fy+1):
            px=np.sum(sh*img[i:i+fx,j:j+fy])
            py=np.sum(sv*img[i:i+fx,j:j+fy])
            res[i,j]=np.sqrt((px*px)+(py*py))
    
    return res.astype(np.uint8)
```

## Line Detection

```python
# Initializing line filters

line_v=[
          [-1,2,-1],
          [-1,2,-1],
          [-1,2,-1]
        ]

line_h=[
          [-1,-1,-1],
          [2,2,2],
          [-1,-1,-1]
        ]
```

```python
# Performing line detection 

edge=conv(gray_im,line_v,line_h)
display(edge,"Line Detection")
```

## Edge Detection

```python
# Performing edge detection using Robert operator

robert_v=[
          [1,0],
          [-1,0]
        ]

robert_h=[
          [1,-1],
          [0,0],
        ]
edge=conv(gray_im,robert_v,robert_h)
display(edge,"Edge Detection using Robert Operator")
```

```python
# Initializing Prewitt Filters

prewitt_v=[
          [1,1,1],
          [0,0,0],
          [-1,-1,-1]
        ]

prewitt_h=[
          [-1,0,1],
          [-1,0,1],
          [-1,0,1]
        ]

# Performing edge detection using Prewitt operator

edge=conv(gray_im,prewitt_v,prewitt_h)
display(edge,"Edge Detection using Prewitt Operator")
```

```python
# Initializing Sobel Filters

sobel_v=[
          [1,2,1],
          [0,0,0],
          [-1,-2,-1]
        ]

sobel_h=[
          [-1,0,1],
          [-2,0,2],
          [-1,0,1]
        ]

# Performing edge detection using sobel operator

edge=conv(gray_im,sobel_v,sobel_h)
display(edge,"Edge Detection using Sobel Operator")
```
